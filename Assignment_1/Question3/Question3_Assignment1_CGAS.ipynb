**Question 3, Part (a)**:

1. **Calculate Ingredient Frequencies**:
   - **Count Frequencies**: Determine how often each ingredient appears across all recipes.

2. **Assign Ranks**:
   - **Rank Ingredients**: Assign ranks based on the frequency of each ingredient, where rank 1 corresponds to the most frequent ingredient.

3. **Normalize Frequencies**:
   - **Convert to Proportions**: Normalize the frequencies to proportions (i.e., divide each frequency by the total count) to get a sense of the relative frequency of each ingredient.

4. **Plot the Frequency-Rank Distribution**:
   - **Create Plot**: Plot the ranks on the x-axis and the normalized frequencies on the y-axis.
   - **Log Scale**: Use a logarithmic scale for both axes to better visualize the distribution of ingredient frequencies (common in rank-frequency plots).
   - **Labeling**: Add titles and labels to make the plot understandable.
   - **Save Plot**: Save the plot in both SVG and PNG formats.

```
import pandas as pd
import matplotlib.pyplot as plt

# Load the extracted ingredients CSV file
df = pd.read_csv('Question1_B.csv')

# Calculate frequency of each ingredient
ingredient_frequencies = df['Ingredient'].value_counts()

# Assign ranks based on frequency
ranks = range(1, len(ingredient_frequencies) + 1)

# Normalize frequencies to get proportions
total_count = ingredient_frequencies.sum()
normalized_frequencies = ingredient_frequencies / total_count

# Plot the frequency-rank distribution
plt.figure(figsize=(10, 6))
plt.plot(ranks, normalized_frequencies.sort_values(ascending=False), marker='o', linestyle='-', color='b')
plt.title('Frequency-Rank Distribution of Ingredients')
plt.xlabel('Rank')
plt.ylabel('Normalized Frequency')
plt.xscale('log')  # Log scale for rank
plt.yscale('log')  # Log scale for frequency
plt.grid(True, which='both', linestyle='--', linewidth=0.7)

# Save the plot as SVG and PNG files
plt.savefig('Question3_A_Frequency-Rank_Distribution.svg')
plt.savefig('Question3_A_Frequency-Rank_Distribution.png')
plt.show()
```
